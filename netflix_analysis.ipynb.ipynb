    Duração mais frequente:

    Qual foi a duração de filme mais frequente na década de 1990?
    Salve essa resposta aproximada como um número inteiro na variável duration.
    Considere 1990 como o ano de início da década.



```python
#filtrando apenas filmes(type == 'Movie') lançados entre 1990 e 1999
movies_90s = netflix_df[
    (netflix_df['type'] == 'Movie') &
    (netflix_df['release_year'] >= 1990) &
    (netflix_df['release_year'] <= 1999)
]
```

```python
#não temos valores nulos na coluna 'duration', caso tivessemos teriamos que limpar
movies_90s = movies_90s.dropna(subset=['duration'])
```

```python
duration = int(movies_90s['duration'].mode()[0])
print("Duração mais frequente nos anos 90: ",duration)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criar histograma das durações dos filmes dos anos 90
plt.figure(figsize=(10, 6))
sns.histplot(data=movies_90s, x='duration', binwidth=5, color='skyblue', kde=False)

# Linha para destacar a moda
plt.axvline(duration, color='red', linestyle='--', label=f'Moda: {duration} min')

plt.title('Distribuição da Duração dos Filmes (1990-1999)')
plt.xlabel('Duração (minutos)')
plt.ylabel('Número de filmes')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
action_90s = netflix_df[
    (netflix_df['type'] == 'Movie') &
    (netflix_df['release_year'] >= 1990) &
    (netflix_df['release_year'] <= 1999) &
    (netflix_df['genre'] == 'Action')
]
```

```python
#contar filmes com menos de 90 minutos
short_movie = action_90s[action_90s['duration'] < 90]
short_movie_count = len(short_movie)
print("Quantidade de filmes curtos de ação na década de 1990: ",short_movie_count)
```

```python
# Contar filmes curtos e não curtos
short_count = short_movie_count
long_count = len(action_90s) - short_movie_count

# Criar gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(
    [short_count, long_count],
    labels=['Curtos (<90 min)', 'Não curtos (≥90 min)'],
    autopct='%1.1f%%',
    startangle=90,
    colors=['lightcoral', 'lightgray']
)
plt.title('Filmes de Ação Curtos vs Não Curtos (1990-1999)')
plt.tight_layout()
plt.show()
```
