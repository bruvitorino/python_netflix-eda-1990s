    Durao mais frequente:

    Qual foi a durao de filme mais frequente na dcada de 1990?
    Salve essa resposta aproximada como um nmero inteiro na varivel duration.
    Considere 1990 como o ano de incio da dcada.



```python
#filtrando apenas filmes(type == 'Movie') lanados entre 1990 e 1999
movies_90s = netflix_df[
    (netflix_df['type'] == 'Movie') &
    (netflix_df['release_year'] >= 1990) &
    (netflix_df['release_year'] <= 1999)
]
```

```python
#no temos valores nulos na coluna 'duration', caso tivessemos teriamos que limpar
movies_90s = movies_90s.dropna(subset=['duration'])
```

```python
duration = int(movies_90s['duration'].mode()[0])
print("Durao mais frequente nos anos 90: ",duration)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criar histograma das duraes dos filmes dos anos 90
plt.figure(figsize=(10, 6))
sns.histplot(data=movies_90s, x='duration', binwidth=5, color='skyblue', kde=False)

# Linha para destacar a moda
plt.axvline(duration, color='red', linestyle='--', label=f'Moda: {duration} min')

plt.title('Distribuio da Durao dos Filmes (1990-1999)')
plt.xlabel('Durao (minutos)')
plt.ylabel('Nmero de filmes')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
action_90s = netflix_df[
    (netflix_df['type'] == 'Movie') &
    (netflix_df['release_year'] >= 1990) &
    (netflix_df['release_year'] <= 1999) &
    (netflix_df['genre'] == 'Action')
]
```

```python
#contar filmes com menos de 90 minutos
short_movie = action_90s[action_90s['duration'] < 90]
short_movie_count = len(short_movie)
print("Quantidade de filmes curtos de ao na dcada de 1990: ",short_movie_count)
```

```python
# Contar filmes curtos e no curtos
short_count = short_movie_count
long_count = len(action_90s) - short_movie_count

# Criar grfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(
    [short_count, long_count],
    labels=['Curtos (<90 min)', 'No curtos (90 min)'],
    autopct='%1.1f%%',
    startangle=90,
    colors=['lightcoral', 'lightgray']
)
plt.title('Filmes de Ao Curtos vs No Curtos (1990-1999)')
plt.tight_layout()
plt.show()
```
